## What is matplotlib

Matplotlib is a plotting library

### Why Matplotlib?
* Built on Numpy arrays (and Python)
* Integrates directly with pandas
* Can create basic or advanced plots
* Simple to use interface (once you get the foundations)

  4. Customise plot
  5. Save/share plot
* Importing Matplot and the 2 ways of plotting
* Plotting data from Numpy arrays
* Plotting data from pandas DataFrames

## Matplotlib example workflow


```python
# 0. Import matplotlib and get it ready for plotting
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. setup plot
fig, ax = plt.subplots(figsize=(10 ,8))  # (width, height)

# 3. Plot the data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("sample_data/sample-plot.png")
```

## Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))
# Plot to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))

# Plot to each different index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000));
```

### Which one should you use? (pyplot vs matplotlib OO method )
* When plotting something quickly, okay to use the pyplot method
* When plotting something more advanced, use the OO method


```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
len(over_50)
```

```python
# Pyplot method
over_50.plot(kind="scatter", x="age", y="chol", c="target");
```

```python
# OO method mixed with pyplot
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter", x="age", y="chol", c="target", ax=ax);
# ax.set_xlim([45, 100]);
```

```python
over_50.head()
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# customize the plot
ax.set(title="Heart Disease and Cholesteral Levels",
       xlabel="Age",
       ylabel="Cholesteral")

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target")

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle="--");
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])

# customize ax0
ax0.set(title="Haert Disease and Cholesterol Levels",
        ylabel="Cholesteral")

# Add legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# Add mean line
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--")

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                       y=over_50["thalach"],
                       c=over_50["target"])

# customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate ")

# Add legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# Add mean line
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--");

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

This plot shows some information about the heart disease dataset

```python
# Customizing the y and x axis limitation 

# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                      cmap="winter")

# customize ax0
ax0.set(title="Haert Disease and Cholesterol Levels",
        ylabel="Cholesteral")

# change the x axis limits
ax0.set_xlim([50, 80])

# change the 
# ax0.set_ylim([50])

# Add legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# Add mean line
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--")

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                       y=over_50["thalach"],
                       c=over_50["target"],
                      cmap="winter")

# customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate ")

# change ax1 x lim
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

# Add legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# Add mean line
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--");

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
