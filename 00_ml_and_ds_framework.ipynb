## Main Types of Machine Learning Problems
* Supervised Learning (data, label)
  * classification
    * "Is this example one thing or another"
    * Binary classification = two options
    * Multi-class classification = more than two options
  * regression
    * "Trying to predict a number" / *continuous number*
    * How much will this house sell for?
    * How many people will buy this app?

* Unsupervised Learning (data, no label)
  * clustering
  * recommendation problem
* Transfer Learning
  * leverage what one machine learning model has learned in another machine learning problem
* Reinforcement Learning
  * involves having a computer program performs some actions within a defined space and rewards it for doing it well or purnish it for doing it poorly
